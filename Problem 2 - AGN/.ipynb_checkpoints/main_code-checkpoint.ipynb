```python
import numpy as np
from numpy import exp,pi,sqrt
import matplotlib.pyplot as plt
from scipy.integrate import quad
from numpy import pi,log, exp, log10
```

```python
#Constants
Msun = 2 * 10**33 #in g
MBH = 10 * Msun #in g
G = 6.674 * 10**-8 #cgs
c = 3 * 10**10 #cm/s
rin = (20 * G * MBH)/c**2 #cgs
rout = (1000 * G * MBH)/c**2 #cgs
Tcor = 3.481 * 10**9 #K
mp = 1.673 * 10**-24 #cgs
me = 9.1094e-28 #cgs
sigmat = 6.652 * 10**-25 #cgs - Thomson
sigmab = 5.6704 * 10**-5 #cgs - SB
eta = 0.01 #accretion efficiency
h = 6.6261 * 10**-27 #cgs
hred = 1.0546 * 10**-27 #cgs
k = 1.3087 * 10**-16 #cgs
gff = 1.2
me = 9.1094*10**-28 #g
Z=1

Ledd = (4 * np.pi * G * MBH * mp * c)/sigmat
M_dot = Ledd/(eta * c**2)
```

```python
# Accretion disk spectrum

def T(R):
    T0 = ((3*G*MBH*M_dot) / (8*pi*rin**3*sigmab))**0.25
    return T0*(rin/R)**(3/4)*(1-sqrt(rin/R))**(1/4)

def BB(R,v):
    const = (8*pi*h)/c**2
    return (const* v**3 * R) / (exp((h*v)/(k*T(R))) -1)

v = np.arange(3,22.01,0.1,dtype=float)

L = np.empty(len(v))

for ind, freq in enumerate(v):
    integral = quad(BB, rin, rout, args=(10**freq))
    L[ind] = integral[0]
    

plt.plot(v, np.log10(L), 'o', ms=1)
plt.xlabel(r'$log(\nu / Hz)$ ')
plt.ylabel(r'$log(L_{\nu}$ / $erg \cdot s^{-1} Hz^{-1}$)')
#plt.ylim(0,25)
#plt.savefig('1.png',dpi=150)
```

```python
taus = np.array([0.01,0.1,1.])

V = (4/3)*pi*rout**3
Bremsstrahlung = []
nes = taus/(sigmat*rout)
for ind,tau in enumerate(taus):
    brems = 6.8e-38*nes[ind]**2*np.power(Tcor,-0.5)*gff*exp(-(h*10**v)/(k*Tcor))
    Bremsstrahlung.append(brems*V)
```

```python
# Bremsstrahlung radiation spectrum

for i in range(len(taus)):
    plt.plot(v, np.log10(Bremsstrahlung[i]), 'o', ms=1, label=r'$\tau=$'+str(taus[i]))
    
plt.legend()
plt.xlabel(r'$log(\nu / Hz)$ ')
plt.ylabel(r'$log(L_{\nu}$ / $erg \cdot s^{-1} Hz^{-1}$)')
plt.ylim(-10,25)
#plt.savefig('2.png',dpi=150)
```

```python
def x(en, en1, y): 
    return en1/(4*y**2 * en) 

def Fc(en, en1, y): 
    if(x(en, en1, y) > 1):
        return 0 
    else:
        return 1 + x(en, en1, y) - 2*x(en, en1, y)**2 + 2*x(en, en1, y)*np.log(x(en, en1, y))
    
def N_y(y): 
    return 4*pi*c**3*(me/(2*pi*k*Tcor))**1.5 * sqrt(1-1/y**2)/y**3 * exp(((me*c**2)/(2*k*Tcor))*(1-1/y**2))#np.exp(-(y*me*c**2)/(k*Tcor)) #This function is fine. Derived this manually, so no mistakes


def nphot(en, ind): 
    return (L[ind]/(4*np.pi*rout**2))/(h*en*c )


def int_y(y, en, en1): 
    return Fc(en, en1, y) * N_y(y)/y**2

def int_v(en, en1, ind): 
    integral_y = quad(int_y, 1, 3, args=(en,en1))
    return nphot(en, ind)/en * integral_y[0]


v1 = np.arange(8, 27, 0.1, dtype=float) 

V = 4/3 * np.pi * rout**3
#nes = np.array([0.1/(sigmat*rout)])
Compton = []
 
for ind0,ne in enumerate(nes): 
    emissivity = np.zeros(v1.shape)
    for ind,freq in enumerate(v1):
        integral_v = quad(int_v, 10**3, 10**22, args=(10**freq, ind))
        emissivity[ind] = integral_v[0] * (3 * c * sigmat * 10**freq)/(16*np.pi) * ne * h * V 
        
    Compton.append(emissivity)
        
    plt.scatter(v1, np.log10(emissivity), s=1) 
```

```python
# Spectrum of comptonized photons

for i in range(len(taus)):
    #plt.plot(v, np.log10(L+Bremsstrahlung[i]), 'o', ms=1, color=colors[i],label=r'$\tau=$'+str(taus[i]))
    plt.plot(v1, np.log10(Compton[i]),'o',ms=1,label=r'$\tau=$'+str(taus[i]))

plt.legend()
plt.xlim(9,25)
plt.ylim(-30,27)
plt.xlabel(r'$log(\nu / Hz)$ ')
plt.ylabel(r'$log(L_{\nu}$ / $erg \cdot s^{-1} Hz^{-1}$)')
#plt.savefig('3.png',dpi=150)
```
